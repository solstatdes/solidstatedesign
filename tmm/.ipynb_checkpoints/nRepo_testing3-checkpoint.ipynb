```python
# dummy config json from database
json = {
  "input":{
    "path":"glass/misc/soda-lime/Rubin-lowiron.yml",
    "layer":"Soda Lime Glass"
  },
  "configuration":"substrate",
  "resolution":100,
  "stack":[
    {
      "path":"main/CdTe/Treharne.yml",
      "layer":"Ag",
      "d":100
    },
    {
      "path":"main/CdS/Treharne.yml",
      "layer":"Ag",
      "d":100
    }
  ],
  "output":{
    "path":"glass/misc/soda-lime/Rubin-lowiron.yml",
    "layer":"air"
  }
}
nRepo = {}
```

```python
# want to get all paths
def get_paths(json, nRepo):
    if json.get('path', None):
        stub = json.get('path')
        if nRepo.get(json.get('path'), None):
            next
        else:
            path = os.path.join(LIBRARY_PATH, stub)
            page_data = db.L(path).grabData()
            nRepo[stub] = page_data
    else:
        for item in json:
            if isinstance(json.get(item, None), dict):
                try: 
                    get_paths(json.get(item), nRepo)
                except:
                    continue
            elif isinstance(json.get(item, None), list):
                for item in json.get(item, None):
                    get_paths(item, nRepo)
            
```

```python
nRepo={}
get_paths(json, nRepo)
nRepo
```

```python
#next you want to build a repository of normalized data
#but before that will find out max and min x values
def get_limits(repo):
    mins = []
    maxs = []
    keys = repo.keys()
    for item in keys:
        mins.append(min(repo.get(item).get('x')))
        maxs.append(max(repo.get(item).get('x')))
        
    return [max(mins), min(maxs)]
        
```

```python
wlRange = get_limits(nRepo)
wlRange
```

```python
from scipy import interpolate

def N_interp(repo, wlRange):
    N = {}
    N['x'] = np.linspace(min(wlRange), max(wlRange), 50)
    for item in repo.keys():
        print item
        n_complex = np.array(repo.get(item).get('n')) + np.array(repo.get(item).get('k'))*1j
        x = np.array(repo.get(item).get('x'))
        f = interpolate.interp1d(x, n_complex)
        N[item] = f(N.get('x'))
        
    return N
```

```python
N = N_interp(nRepo, wlRange)
```
